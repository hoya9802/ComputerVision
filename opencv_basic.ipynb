```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
myArray = np.random.rand(3, 2, 4)
print('myArray:\n', myArray)
print()

print('split array0:\n', myArray[:, :, 0])
print()

zeroTxT = np.zeros((10, 10))
print('10x10 ones array')
print(zeroTxT)
print()

onesTxT = np.ones((10, 10))
print('10x10 ones array')
print(onesTxT)
print()

ones100 = onesTxT.reshape(10*10)
print("1x100 ones array")
print(ones100)
print()

print('sum of 10x10 ones array(sum):', sum(onesTxT))
print('sum of 10x10 ones array(np.sum):', np.sum(onesTxT))
print('sum of 1x100 ones array(sum):', sum(onesTxT.reshape(10*10)))
```

```python
img = cv.imread('/content/drive/MyDrive/Computer_Vision/data/color_map.png')
if img is None:
  print('File not found, \nCheck the file name and try again.')
else:
  print(img.shape)
  print(np.sum(img))
  cv2_imshow(img)
```

```python
# print Blue color
blue_img = img[:,:,0]
print('blue color array:\n', blue_img)
print('Sum of total blue color pixel(np):', np.sum(blue_img))
print('Sum of total blue color pixel:', sum(blue_img.reshape(100)))

print()

# print Green color
green_img = img[:,:,1]
print('green color array:\n', green_img)
print('Sum of total green color pixel(np):', np.sum(green_img))
print('Sum of total green color pixel:', sum(green_img.reshape(100)))

print()

# print Red color
red_img = img[:,:,2]
print('red color array:\n', red_img)
print('Sum of total red color pixel(np):', np.sum(red_img))
print('Sum of total red color pixel:', sum(red_img.reshape(100)))
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
print(gray_img)
print('sum of gray image pixels:', np.sum(gray_img))
cv2_imshow(gray_img)
```

```python
upsampled_color = cv.resize(img, (100, 100))
cv2_imshow(upsampled_color)
print('sum of upsampled color image:', sum(upsampled_color.reshape(30000)))
print('sum of upsampled color image(np):', np.sum(upsampled_color))

upsampled_gray = cv.resize(gray_img, (100, 100))
cv2_imshow(upsampled_gray)
print('sum of upsampled color image:', sum(upsampled_gray.reshape(10000)))
print('sum of upsampled color image(np):', np.sum(upsampled_gray))
```
